```python
from bs4 import BeautifulSoup

def parse_html_by_header(html_string):
    soup = BeautifulSoup(html_string, 'html.parser')
    current_section = None
    seen_texts = set()

    for element in soup.descendants:
        if not element.string:
            continue
        text = element.string.strip().replace('\xa0', ' ')
        if text in seen_texts:
            continue

        if element.name == 'h2':
            current_section = {'header': text, 'content': []}
            seen_texts.add(text)
        elif element.name == 'a':
            current_section['content'].append({
                'text': text,
                'link': element.get('href')
            })
            seen_texts.add(text)
        elif element.name is None:
            current_section['content'].append({
                'text': text,
                'link': None
            })
            seen_texts.add(text)

    return current_section

for i in range(len(dataset)):
    html_content = dataset["Extractor 1 1"][i]

    section = parse_html_by_header(html_content)
    if section['header'] == "Что я могу узнать о заказчике с помощью Аналитики?":
        print(section)
```
